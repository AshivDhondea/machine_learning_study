# Iris dataset - Exploratory Data Visualization

## Dataset information
[Kaggle source for dataset and background information](https://www.kaggle.com/datasets/uciml/iris/data)

>The Iris dataset was used in R.A. Fisher's classic 1936 paper, The Use of Multiple Measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning Repository.

>It includes three iris species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.

>The columns in this dataset are:

    > Id
    > SepalLengthCm
    > SepalWidthCm
    > PetalLengthCm
    > PetalWidthCm
    > Species

This notebook draws inspiration from [iris Flower EDA notebook on Kaggle](https://www.kaggle.com/code/abdmental01/iris-flower-eda-ml-beginner) and [python data visualizations notebook on Kaggle](https://www.kaggle.com/code/benhamner/python-data-visualizations)

```python
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
# Check if file path is valid.
file_path = './datasets/iris-species/Iris.csv'
if not os.path.isfile(file_path):
    fnf_err = f'{file_path} not found.'
    raise FileNotFoundError(fnf_err)

# Read dataset into a pandas dataframe.
df = pd.read_csv(file_path)

# Take a look at the first 5 rows of the dataset.
print(df.head(5))
# Look at column names and the shape of the dataset.
columns = list(df.columns)
print(f"The columns are {columns}")
num_rows, num_cols = df.shape
print(f"There are {num_rows} rows and {num_cols} columns.")
    
```

SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm are quantitative (numerical) data.
Species is qualitative (categorical) data.

```python
df_null = df.isnull().sum().sum()
print(f'Checking if there are any null entries in the dataframe: {df_null}')

# Count duplicate rows in train_data
data_duplicates = df.duplicated()
# Print the results
print(f"Duplicate rows in the dataframe: {data_duplicates}")

# For the categorical variable, count the frequency of each category
print(df["Species"].value_counts())

# For the numerical variables, describe the statistics.
print(df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]].describe())
```

```python
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
# We can set the number of bins with the *bins* keyword argument.
n_bins = 15
axs[0].hist(df['SepalLengthCm'], bins=n_bins, color='c', edgecolor='k')
axs[1].hist(df['SepalWidthCm'], bins=n_bins, color='c', edgecolor='k')
axs[0].axvline(df['SepalLengthCm'].mean(), color='r', linestyle='solid', linewidth=1)
axs[0].axvline(df['SepalLengthCm'].quantile(q=0.25), color='b', linestyle='dotted', linewidth=1)
axs[0].axvline(df['SepalLengthCm'].quantile(q=0.75), color='orange', linestyle='dashed', linewidth=1)
axs[0].axvline(df['SepalLengthCm'].median(), color='pink', linestyle='dashdot', linewidth=1)
axs[1].axvline(df['SepalWidthCm'].mean(), color='r', linestyle='solid', linewidth=1)
axs[1].axvline(df['SepalWidthCm'].quantile(q=0.25), color='b', linestyle='dotted', linewidth=1)
axs[1].axvline(df['SepalWidthCm'].quantile(q=0.75), color='orange', linestyle='dashed', linewidth=1)
axs[1].axvline(df['SepalWidthCm'].median(), color='pink', linestyle='dashdot', linewidth=1)
axs[0].set_xlabel('Sepal Length [cm]')
axs[1].set_xlabel('Sepal Width [cm]')
axs[0].set_ylabel('Count')
plt.show()
```

```python
fig1, axs1 = plt.subplots(1, 2, sharey=True, tight_layout=True)
axs1[0].hist(df['PetalLengthCm'], bins=n_bins, color='c', edgecolor='k')
axs1[1].hist(df['PetalWidthCm'], bins=n_bins, color='c', edgecolor='k')
axs1[0].axvline(df['PetalLengthCm'].mean(), color='r', linestyle='solid', linewidth=1)
axs1[0].axvline(df['PetalLengthCm'].quantile(q=0.25), color='b', linestyle='dotted', linewidth=1)
axs1[0].axvline(df['PetalLengthCm'].quantile(q=0.75), color='orange', linestyle='dashed', linewidth=1)
axs1[0].axvline(df['PetalLengthCm'].median(), color='pink', linestyle='dashdot', linewidth=1)
axs1[1].axvline(df['PetalWidthCm'].mean(), color='r', linestyle='solid', linewidth=1)
axs1[1].axvline(df['PetalWidthCm'].quantile(q=0.25), color='b', linestyle='dotted', linewidth=1)
axs1[1].axvline(df['PetalWidthCm'].quantile(q=0.75), color='orange', linestyle='dashed', linewidth=1)
axs1[1].axvline(df['PetalWidthCm'].median(), color='pink', linestyle='dashdot', linewidth=1)
axs1[0].set_xlabel('Petal Length [cm]')
axs1[1].set_xlabel('Petal Width [cm]')
axs1[0].set_ylabel('Count')
plt.show()
```

```python
fig2, axs2 = plt.subplots(2, 1, figsize=(5, 10), sharex=False, sharey=False,
                            tight_layout=True)
axs2[0].hist2d(df['SepalLengthCm'], df['SepalWidthCm'], bins=15)
axs2[0].set_xlabel('Sepal Length [cm]')
axs2[0].set_ylabel('Sepal Width [cm]')
axs2[0].axvline(df['SepalLengthCm'].mean(), color='r', linestyle='solid', linewidth=1)
axs2[0].axhline(df['SepalWidthCm'].mean(), color='m', linestyle='dashed', linewidth=1)

axs2[1].hist2d(df['PetalLengthCm'], df['PetalWidthCm'], bins=15)
axs2[1].axvline(df['PetalLengthCm'].mean(), color='r', linestyle='solid', linewidth=1)
axs2[1].axhline(df['PetalWidthCm'].mean(), color='m', linestyle='dashed', linewidth=1)
axs2[1].set_xlabel('Petal Length [cm]')
axs2[1].set_ylabel('Petal Width [cm]')
plt.show()
```

```python
fig3, axs3 = plt.subplots(1, 4, figsize=(20, 5), tight_layout=True)
# plot violin plot
axs3[0].violinplot(df['SepalLengthCm'],showmeans=True, showmedians=False)
axs3[0].set_title('Sepal Length [cm]')
axs3[1].violinplot(df['SepalWidthCm'],showmeans=True, showmedians=False)
axs3[1].set_title('Sepal Width [cm]')
axs3[2].violinplot(df['PetalLengthCm'], showmeans=True, showmedians=False)
axs3[2].set_title('Petal Length [cm]')
axs3[3].violinplot(df['PetalWidthCm'], showmeans=True, showmedians=False)
axs3[3].set_title('Petal Width [cm]')
# adding horizontal grid lines
for ax in axs3:
    ax.yaxis.grid(True)
    ax.set_ylabel('[cm]')
plt.show()
```
